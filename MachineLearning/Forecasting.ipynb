```python
import pandas as pd
import statsmodels.api as sm
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.preprocessing import MinMaxScaler, RobustScaler
from sklearn.metrics import r2_score, mean_squared_error

%matplotlib inline  

url = 'https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/datasets/AirPassengers.csv'
airPass = pd.read_csv(url)
y = pd.Series(airPass['value'])
y.index = pd.date_range(start='1949-1-1', periods = len(y), freq='m')
print(y.head())
```

```python
plt.figure()
y.plot()
plt.title('AirPassenger Data')
plt.show()
```
